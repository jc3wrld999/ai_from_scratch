- gradient descent:

```python
import numpy as np

#  1   

def linear_model(w0, w1, X):
    
    f_x = w0 + w1 * X
    
    return f_x
    
'''
1.   ' '   
   MSE   . 
'''

def Loss(f_x, y):
    
    ls = np.mean(np.square(y - f_x))
    
    return ls

'''
2.   'Gradient'       
   w0 w1  gradient 'gradient0' 'gradient1'
     gradient_descent  .
   
   Step01. w0  gradient 'gradient0' .
   
   Step02. w1  gradient 'gradient1' .
'''

def gradient_descent(w0, w1, X, y):
    
    gradient0 = 2 * np.mean((y - (w0 + w1 * X)) * -1)
    gradient1 = 2 * np.mean((y - (w0 + w1 * X)) * -1 * X)
    
    return np.array([gradient0, gradient1])

'''
3.   ' '     
   gradient descent     .
   
   Step01.   gradient_descent  
           w0 w1  gradient 'gd' .
           
   Step02.  'w0' 'w1'   w0 w1 
             .  
            'gd'    'lr' .
'''

def main():
    
    X = np.array([1,2,3,4]).reshape((-1,1))
    y = np.array([3.1, 4.9, 7.2, 8.9]).reshape((-1,1))
    
    #  
    w0 = 0
    w1 = 0
    
    # learning rate 
    lr = 0.001
    
    #   1000 
    for i in range(1000):
    
        gd = gradient_descent(w0, w1, X, y)
        
        w0 = w0 - lr * gd[0]
        w1 = w1 - lr * gd[1]
        
        # 100  loss w0, w1 
        if (i % 100 == 0):
        
            loss = Loss(linear_model(w0,w1,X),y)
        
            print("{} loss : {}".format(i, loss))
            print("{} w0, w1 : {}, {}".format(i, w0, w1),'\n')

    return w0, w1

if __name__ == '__main__':
    main()
```
