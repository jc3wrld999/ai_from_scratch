## 데이터 시각화




```python
import seaborn as sns
sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins').dropna()
penguins
```

## Lineplot
- 선그래프: 특정 데이터를 x, y로 표시하여 관계를 확인
- 수치 데이터 확인

```python
# ci: 오차 영역
# hue: species(종)에 따라 다른색
sns.lineplot(data=penguins, x='body_mass_g', y='flipper_length_mm', ci=None, hue='species') 
```

## 크기 표현
- Pointplot
- Histplot
- Barplot
- Countplot

```python
# palette는 색깔/tip 모르면 palette='a'로 에러코드에서 찾으면 됨
sns.pointplot(data=penguins, x='species', y='body_mass_g', hue='sex', palette='winter')
```

### Histplot
- 연속된 데이터의 특정 구간별 정보 확인

```python
sns.histplot(data=penguins, x='bill_depth_mm', bins=30, hue='species', multiple='stack')
```

### Barplot
- 막대그래프
- 히스토그램과 다름
- 가로/세로 둘다 가능

```python
sns.barplot(data=penguins, x='island', y='body_mass_g', hue='sex')
```

### Countplot
- 범주형 속성을 가지는 데이터들의 히스토그램
- 종류별 count를 보여줌

```python
sns.countplot(data=penguins, x='species')
```

## 분포 표현하기
- Boxplot

### Boxplot
- 데이터의 각 종류별로 사분위 수를 표현
- 특정 데이터의 전체적인 분포를 확인하기 좋음

```python
sns.boxplot(data=penguins, x='species', y='body_mass_g')
```
