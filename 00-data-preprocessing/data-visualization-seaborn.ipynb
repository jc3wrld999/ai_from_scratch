##  




```python
import seaborn as sns
sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins').dropna()
penguins
```

## Lineplot
- :   x, y   
-   

```python
# ci:  
# hue: species()  
sns.lineplot(data=penguins, x='body_mass_g', y='flipper_length_mm', ci=None, hue='species') 
```

##  
- Pointplot
- Histplot
- Barplot
- Countplot

```python
# palette /tip  palette='a'   
sns.pointplot(data=penguins, x='species', y='body_mass_g', hue='sex', palette='winter')
```

### Histplot
-      

```python
sns.histplot(data=penguins, x='bill_depth_mm', bins=30, hue='species', multiple='stack')
```

### Barplot
- 
-  
- /  

```python
sns.barplot(data=penguins, x='island', y='body_mass_g', hue='sex')
```

### Countplot
-     
-  count 

```python
sns.countplot(data=penguins, x='species')
```

##  
- Boxplot

### Boxplot
-      
-      

```python
sns.boxplot(data=penguins, x='species', y='body_mass_g')
```
