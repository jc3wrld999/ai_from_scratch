Bag of words는 단어의 빈도수를 기반으로 문서 벡터를 생성합니다.


이번 실습에서는 scikit-learn의 CountVectorizer를 사용하여, bag of words 문서 벡터를 만들어 보는 실습을 진행하겠습니다.

영화 리뷰 데이터인 text.txt에 저장되어 있는 IMDB dataset을 사용하여 각 리뷰별 문서 벡터를 만들어 보세요.

```python
import re
from sklearn.feature_extraction.text import CountVectorizer

regex = re.compile('[^a-z ]')


with open("text.txt", 'r', encoding='UTF8') as f:
    documents = []
    for line in f:
        # doucments 리스트에 리뷰 데이터를 저장하세요.
        documents.append(regex.sub("", line.rstrip()))
        
# CountVectorizer() 객체를 이용해 Bag of words 문서 벡터 생성
cv = CountVectorizer()
X = cv.fit_transform(documents)

print(f'X의 차원\n{X.shape}\n')

words_feature = cv.get_feature_names()[:10]
print(f'생성한 CountVectorizer() 객체에서 첫 10개의 칼럼이 의미하는 단어\n{words_feature}\n')

idx = cv.vocabulary_['comedy']
print(f'"comedy"를 의미하는 칼럼의 인덱스 값\n{idx}\n')

vec1 = X[0]
print(f'첫 번째 문서의 Bag of words 벡터:\n{vec1}\n')
```
