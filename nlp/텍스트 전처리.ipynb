## Stopword 제거
- 문법적 기능을 지닌 단어 및 불필요하게 자주 발생하는 단어 제거
```python
import nltk # nltk 라이브러리 안에 stopwords가 저장되있음
from nltk.corpus import stopwords
sentence = ['the', 'green', 'egg', 'and', 'ham', 'a', 'an']
stopwords = stopwords.words('english')
new_sentence = [word for word in sentence if word not in stopwords] # sentence에서 stopwords에 포함되지 않은 단어만 추출
print(new_sentence)
stopwords += "승주" # 이런 식으로 추가도 가능
```

## Stemming
- 동일한 의미의 단어지만 문법적인 이유 등 표현 방식이 다양한 단어를 공통된 형태로 변환
```python
import nltk
from nltk.stem import PorterStemmer
words = ['studies', 'studied', 'studying', 'dogs', 'dog']
stemmer = PorterStemmer()
for word in words:
    print(stemmer.stem(word)) # studi, studi, studi, dog, dog
```

````python
word_counter = dict()

# 단어가 key, 빈도수가 value로 구성된 딕셔너리 생성
with open('text.txt', 'r', encoding='UTF8') as f:
    for line in f:
        for word in line.rstrip().split():
            if word not in word_counter:
                word_counter[word] = 1
            else:
                word_counter[word] += 1

print(word_counter)


# 텍스트 파일에 내 모든 단어의 총 빈도수
total = sum(word_counter.values())

# 텍스트 파일 내 100회 이상 발생하는 단어를 리스트 형태로 저장
up_five = list(key for key, value in word_counter.items() if value >= 100)


print(total)
print(up_five)
````

````python
import re

word_counter = dict()
regex = re.compile('[^a-zA-Z]')

# 텍스트 파일을 소문자로 변환 및 숫자 및 특수기호를 제거한 딕셔너리 생성
with open('text.txt', 'r', encoding='UTF8') as f:
    for line in f:
        for word in line.rstrip().split():
            word = regex.sub('', word.lower())
            if word not in word_counter:
                word_counter[word] = 1
            else:       
                word_counter[word] += 1

# 단어 "the"의 빈도수
count = word_counter["the"]

print(count)
````
